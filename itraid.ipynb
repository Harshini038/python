```python
class Politician:
    def __init__(self, partyname):
        self.partyname = partyname
        
    def getLIMIT(self):
        return 50000000
    
    def doITRaid(self):
        if self.hasProperty():
            self.evaluateProperty()
        else:
            return 0
        
    
class Professional:
    def __init__(self, qual):
        self.qualification = qual
        
    def getLIMIT(self):
        return 10 * self.monthly_income * 12
    
    def doITRaid(self):
        if(self.hasDeposit()):
            return self.getDeposit()
        else:
            return 0
        
        
class Person:
    def __init__(self,name,status):
        self.__name=name
        self.__status=status
        
    @property              #for getting status
    def status(self):
        return self.__status
    
    @status.setter         #for setting status
    def status(self,status):
         self.__status = status
    
    
    def ITRaid(self):
        value=self.doITRaid()
        if (value>self.getLIMIT()):
            if(self.status=='Minister'):
                raise ITRaid("you hav disproportionate assets "+str(value))
            else:
                raise ITRaid("your deposits are locked "+str(value))
        return value
    

class Minister(Person,Politician):
    
    def __init__(self, name, party_name, asset_value):
        self.asset_value = asset_value
        Person.__init__(self, name, 'Minister')
        Politician.__init__(self, party_name)
    
    def hasProperty(self):
        return bool(self.asset_value)
    
    def evaluateProperty(self):
        return sum(self.asset_value.values())
    
    def __str__(self):
        return "Name: " + self.name 
    

    
class Officer(Person,Professional):
    
     def __init__(self, name, qualification, monthly_income, deposit_value):
        self.monthly_income = monthly_income
        self.deposit_value = deposit_value
        Person.__init__(self, name, 'Officer')
        Professional.__init__(self, qualification)
    
        
     def hasDeposit(self):
        return bool(self.deposit_value)
    
     def getDeposit(self):
        return sum(self.deposit_value.values())\
    
     def __str__(self):
        return "Name: "+self.name
    
    
    
class ITRaid(Exception):
    def __init__(self,errmsg):
        self.errmsg=errmsg
        
    def __str__(self):
        return "ITraid exception "+self.errmsg
    
    
    
        
```
